```python
# List of specific keys to remove
specific_keys_to_remove = ['created-at', 'landed-at', 'accepted-site-id', 'last-seen', 'clicked-at', 'visited-at', 'accepted-at', 'last-visit-at', 'first-seen']#, 'items']

# Recursive function to remove specified entries and 10-digit numbers

def remove_entries(d):
    if isinstance(d, dict):
        return {k: remove_entries(v) for k, v in d.items() if not (k == 'items' and v == []) and k not in specific_keys_to_remove and not (isinstance(v, int) and re.match(r'^\d{10}$', str(v)))}
    elif isinstance(d, list):
        return [remove_entries(item) for item in d if not (isinstance(item, int) and re.match(r'^\d{10}$', str(item)))]
    else:
        return d
# Apply the function to each cell in the DataFrame
```

```python
def drop_links(text):
        text = str(text).lower()
        text = re.sub(r"\n", " ", text)
        text=re.sub(r'[^\sa-zA-Z0-9@]',' ',text)# Удаляем специальные знаки
        text=re.sub(r'[\]\[]',' ',text)
        return re.sub(" +", " ", text)
```

```python
# Cleare Text
ds['features'] = ds['features'].apply(remove_entries)
ds['features'] = ds['features'].map(drop_links)
dsv['features'] = dsv['features'].apply(remove_entries)
dsv['features'] = dsv['features'].map(drop_links)
dst['features'] = dst['features'].apply(remove_entries)
dst['features'] = dst['features'].map(drop_links)
```
